# Who is the greatest baseball player of all time? #
In the early 2000's, a group of 19 sabermetricians got together and tried to answer the most difficult baseball question they could. *Who Is the Greatest Baseball Player of All Time?* They listed 100 possible players and through some process known only to them, they all arrived at the same answer. __Babe Ruth__

They reached this conclusion before another baseball great, Barry Bonds, really started to show what we could do. So, we now have a new debate. *Is Babe Ruth or Barry Bonds the Greatest Baseball Player of All Time?*

If you look at the career stats for both players, you can see that they're fairly comparable.

Ruth played the first part of his career in the dead-ball era, which was a time when offense was very low. The ball had a rubber core, which is softer than the cork center used today. The same ball was used for the whole game and became softer and harder to hit as the game progressed. Ball parks were also larger then than they are today. ## Reasons why we can't just compare the numbers ##
1. They played at different times. The game was different. Dead-ball era vs. offensive era.
2. The level of competition was different. 
  - There were no Latin or African-American players when Ruth played. 
  - The population of the US was smaller overall when Ruth played.
  - Players were drafted and served in WWI, which affected competitiveness of the league
3. The equipment, training, nutrition, injury prevention and recovery were different.
  - Ruth was overweight and drank too much. Today, maybe that wouldn't be allowed, as much.
  - Bonds had an injury in 1999 that would have ended his career if he played in Ruth's day

## Approaches to normalizing their stats ##
One approach to comparing the players is to put their stats on in a neutral environment, and then compare them in that environment. This isn't just a neutral park, but a neutral time and run-scoring environment.

1. The level of offense has varied greatly over time. Size of the strike zone, height of pitching mound, size of stadium change offensive stats for that time. Example: In 1968, the AL average BA was .230. In 1930, the NL average BA was .303. A player with a .301 BA in 1930 was very different than a player with a BA of .301 in 1968. 
2. 1985 American League rates as the most average of all time. How would Bonds and Ruth perform if they both played in a neutral park in the 1985 AL?
3. Normalize each player's stats
- Evaluate player's performance in terms of standard deviations about the average at his time, adjusting for park factors
  - Translate into 1985 avg using standard deviations at player's time. 
  - Example: 
    - Ruth hit .356 in 1927 AL, .360 with park factors considered. 
    - 1927 AL had league avg of .293, with std of .044. 
    - A .360 avg is 1.52 std dev from mean. 
    - In 1985, AL had BA of .261 with std dev of .036. 
    - A BA that is 1.52 std dev is .316. 
    - Therefore, Ruth's BA in neutral 1985 is .316.
4. Compare players' normalized stats

| Player | Avg | OBP | SLG | N_Avg | N_OBP | N_SLG |
|---|---|---|---|---|---|---|
| Ruth | .344 | .471 | .694 | .312 | .436 | .664 |
| Bonds | .300 | .443 | .611 | .299 | .432 | .599 |

5. This result tells us that Ruth was the better hitter relative to his time. We're comparing how many std above the mean a hitter at the time that he played.

## Competition level ##
Comparing Ruth and Bonds to their peers doesn't account for the level of competition that they each saw, and how they would do in a different competitive time. Some facts about baseball today and back in the day.
* The average baseball player is 1/2 inch taller today than 50 years ago. Points to better nutrition.
* In 1964, about 4 percent of MLB PA were by players over 35. Today, it's about 14 percent. Players are playing longer. Points to improvements in training. 
* In 2005, there were 22 pitchers on opening day rosters who had undergone Tommy John surgery, which was invented in 1974. That's about 6 percent of active MLB pitchers, and includes some stars. Their careers would have ended if they played prior to 1974. 
* Minor leagues started in 1930s, which led to better training of players.

### Competition rating ###
We can compare players season-to-season, or environment-to-environment to determine how the competition in the seasons or environments compare to each other. We use a player's Equivalent Average (EqA), which is a league and park independent measure of a player's offensive production. Normalized every year so that average player has EqA of .260. An EqA over .300 is very good.

__Example:__ Group of players in Triple-A Minor Leagues could have .290 EqA, and then a .270 EqA in the Majors. The ratio of those tells us how much more difficult the Majors are than the Minors. 

__Example:__ Group of MLB players in 1943 and 1944 and see that the 1944 players had an EqA 5 percent greater than in 1943. Conclude that the league was 5 percent easier in 1944 than in 1943. 

Conclusion from comparing EqA season-to-season is that the competition has gotten tougher over time. We can use this to place a player from one era in the competitive environment of another era. 

__Example:__ In 1941, Ted Williams became last player to finish a season with a BA over .400. He had a .417 EqA in a league with a difficulty rating of .969. The 2005 AL had a difficulty rating of 1.159. Placing Williams in the 2005 AL environment, he would have an EqA of $$ .417 * \frac{.969}{1.159}=.349 $$ 

If we use this approach for Ruth, we see that most of his games occurred in leagues with difficulty ratings of 0.76 - 0.86, which are below average. The average league difficulty over time is 1.0 and it happened in 1975. Ruth's career EqA is .373, and when adjusted for difficulty rating, it drops to .274. That .274 makes him an above-average player, but barely an All-Star and certainly not a Hall of Famer. 

Bonds is far superior to Ruth when we approach the problem this way. His EqA is around .340. 

But, is this a fair comparison? Ruth is being judged by modern competition levels without the opportunity to get modern training, equipment, playing conditions, etc. If he were playing today, he would have grown up in today's system, not the system from 1900.

#### Other factors to consider ####
* Bonds was a better fielder than Ruth. But, Ruth was a pitcher for the first half of his career.
* Bonds was a better baserunner than Ruth, back when people cared about things like stolen bases.
* Bonds may have had "help" to produce his numbers.
