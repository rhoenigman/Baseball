# Old player's skills, young player's skills #

This is the story of Kevin Maas. Who? Exactly.

Based on, "Why Was Kevin Maas a Bust?", by Nate Silver.

### Who was Kevin Maas? ###
In 1990, the NY Yankees were really bad. Their first baseman, Don Mattingly, was hurt and they brought up minor leaguer Kevin Maas to fill in. Maas was old for a minor leaguer, 25, and he had bad knees. He was sub-par defensively and struck out 20\% of the time. But, he could hit the ball hard and draw walks. 

When he joined the Yankeees in July, he hit 10 HR in his first 77 AB, making him the fastest player in MLB history to hit 10 HR. He was a star, and many sports writers and players called him "the real thing". He finished the season with 21 HR even though he only played half the season.

The next season, he hit 8 HR before Memorial Day, but then went into a slump. He finished the season with a BA of .220 as pitchers learned how to exploit his weaknesses. In 1992, his struggles continued and he was moved back to the Minor Leagues. At the start of the 1993 season, the Yankees released him. He was out of baseball.


### Old player's skills and young player's skills ###
In the 1990s, the use of analytics in baseball was not what it is today. The argument around Maas was the no one could have predicted his rapid rise, or his rapid decline. Maybe even with today's analytics it couldn't be predicted. Who knows? 

In 1987, Bill James introduced the idea of __Old player's skills__ and __Young player's skills.__ James argued that players lose speed and batting average as they age and they learn to rely more on power and walks. Speed and batting average are young player's skills. Power and walks are old player's skills, and young players with old player's skills will deteriorate more quickly, which ultimately results in shorter careers. Kevin Maas possessed old player's skills. He was slow and had a lower batting average. But, he could hit for power and draw walks.

### Criteria for comparison ###
Establish a baseline of performance over a few years for set of variables: Power, walks, body type, speed. Use baseline and player's age to predict future performance. 

The first question: How consistent are common statistics for measuring year-to-year performance for a player. 

| Stat | Corr Coef |
|---|---|
| BA | 0.43 |
| ISO | 0.76 |
| BB | 0.76 |
| SO | 0.83 |
| SB | 0.81 |


For ISO, BB, SO, and SB, the correlations are pretty high. However, for BA, that's not the case, which tells us that BA is much more subject to randomness than the other stats and less predictive of year-to-year performance. 

### Regression to the mean ###
Regression to the mean means that a value will move away from extremes and towards a baseline value. You'll hear this term a lot in life.

* Players with an unusually high BA are likely to see their BA decline in the next season. 

* Players with a low BA and a low SO rate are likely to see an increase in BA the next season. They've been unlucky.

* Players with many 2B and few HR is likely to see an increase in HR in the following season and a decrease in 2B.

* Players with high BB rate, but low ISO, can expect to see an increase in ISO in next season. Players tend to hit for power in deeper counts.

* Coversely, players with high HR can expect pitchers to be more careful and they will get more BB.

### Player types and career path ###
In 1982, Bill James evaluated the offensive statistics for thousands of baseball players and found that the greatest number of players peak at age 27. Until that time, most people thought that players peaked between 28-32. 

James's finding had implications for free agency and teams evaluating free agent talent. Players can't become free agents until they've been in the league for 6 years, which typically happens between ages 28-30. Teams signing a 29-year-old free agent have to be leary of what they're getting. How quickly is that player going to decline? 

James produced a graph showing average EqR vs. Age for all players he evaluated. 

### Career path categories ###
James's graph shows the average EqR. But, coming back to our original questions of categories of players and old player's and young player's skills, Are there also categories of decline for players? Can we expect a player with a certain type of skill set to decline faster than another player with a different skill set? 

Assume three hypothetical players:
* Eddie Early - best year at age 23
* Pete Peaky - valuable between 24-29, but not before or after
* Len Latebloom - best year at age 31.

There are plenty of players who fit these three categories. James isn't saying that these are the three categories, only that there are players who fit these descriptions. 

The average career trajectory doesn't predict very well how these players will perform as their careers progress. 

### Multiple skills ###
Players who do multiple things well tend to age better than players who only do one thing well. For example, players with both speed and power will age better than players with just speed or just power.
